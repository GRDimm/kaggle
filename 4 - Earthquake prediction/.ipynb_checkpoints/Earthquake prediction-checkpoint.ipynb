```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import lightgbm as lgb
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, mean_squared_error

class BaseModel:
    def __init__(self, model, segment_length=150000, rolling_windows=[5, 20, 100, 250, 500, 1000, 5000]):
        self.model = model
        self.segment_length = segment_length
        self.rolling_windows = rolling_windows

    def _prepare_features(self, df):
        df = df.copy().reset_index(drop=True)
        # Compute EMAs and Volatilities for specified rolling windows
        X = pd.DataFrame()
        X['acoustic_data'] = df['acoustic_data'].values

        for window in tqdm(self.rolling_windows):
            X[f'ema_{window}'] = df['acoustic_data'].ewm(span=window, adjust=False).mean()
            X[f'vol_{window}'] = df['acoustic_data'].rolling(window=window).std().fillna(0)

        # Drop NaN values introduced by rolling windows
        X = X.dropna().reset_index(drop=True)
        return X

    def train(self, df):
        X = self._prepare_features(df)
        y = df['time_to_failure'].iloc[len(df) - len(X):].values  # Align target with features
        print("Fitting model")
        self.model.fit(X, y)
        print("Finished to fit model")

    def predict(self, df):
        X = self._prepare_features(df)
        return self.model.predict(X)

    def evaluate(self, df):
        print("Evaluating model")
        X = self._prepare_features(df)
        y_true = df['time_to_failure'].iloc[len(df) - len(X):].values
        y_pred = self.model.predict(X)

        mae = mean_absolute_error(y_true, y_pred)
        rmse = np.sqrt(mean_squared_error(y_true, y_pred))
        return mae, rmse

class LinearRegressionModel(BaseModel):
    def __init__(self, segment_length=150000, rolling_windows=[5, 20, 100, 250, 500, 1000, 5000]):
        super().__init__(LinearRegression(), segment_length, rolling_windows)

class LightGBMModel(BaseModel):
    def __init__(self, segment_length=150000, rolling_windows=[5, 20, 100, 250, 500, 1000, 5000]):
        super().__init__(lgb.LGBMRegressor(), segment_length, rolling_windows)

class XGBoostModel(BaseModel):
    def __init__(self, segment_length=150000, rolling_windows=[5, 20, 100, 250, 500, 1000, 5000]):
        super().__init__(xgb.XGBRegressor(objective='reg:squarederror'), segment_length, rolling_windows)

class RandomForestModel(BaseModel):
    def __init__(self, segment_length=150000, rolling_windows=[5, 20, 100, 250, 500, 1000, 5000]):
        super().__init__(RandomForestRegressor(), segment_length, rolling_windows)

# model = LinearRegressionModel()
model = LightGBMModel()
# xgboost_model = XGBoostModel()
# random_forest_model = RandomForestModel()

mid = len(df) // 2
train_df = df.iloc[:mid]
test_df = df.iloc[mid:]

model.train(train_df)

#predictions = linear_model.predict(test_df)

mae, rmse = model.evaluate(train_df)
print("Train : ", "MAE:", mae, "RMSE:", rmse)

mae, rmse = model.evaluate(test_df)
print("Test : ", "MAE:", mae, "RMSE:", rmse)
```

```python

acoustic_data = df["acoustic_data"]

# Définir les fenêtres
windows = [5, 10, 20, 50, 100, 200, 500]

# Calcul des SMAs (Simple Moving Averages)
for window in windows:
    df[f"SMA_{window}"] = acoustic_data.rolling(window=window).mean()

# Calcul des EMAs (Exponential Moving Averages)
for window in windows:
    df[f"EMA_{window}"] = acoustic_data.ewm(span=window, adjust=False).mean()

# Calcul des écarts-types mobiles
for window in windows:
    df[f"STD_{window}"] = acoustic_data.rolling(window=window).std()

# Calcul des bandes de Bollinger pour une fenêtre de 20 périodes et 2 écarts-types
bollinger_window = 20
bollinger_std = 2

df["Bollinger_Mean"] = acoustic_data.rolling(window=bollinger_window).mean()
df["Bollinger_Upper"] = df["Bollinger_Mean"] + (df["STD_20"] * bollinger_std)
df["Bollinger_Lower"] = df["Bollinger_Mean"] - (df["STD_20"] * bollinger_std)

df.dropna(inplace=True)
```
